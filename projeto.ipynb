```python
import numpy as np
import random

# Definições do problema
disciplinas = ["Matemática", "Português", "História"]
professores = ["Prof_A", "Prof_B", "Prof_C"]
horarios = ["07:00", "08:00", "09:00"]
turmas = ["Turma_1", "Turma_2"]
dias = ["Segunda", "Terça", "Quarta"]

# Carga horária semanal das disciplinas (total de aulas por semana)
carga_horaria_disciplina = {
    "Matemática": 4,
    "Português": 3,
    "História": 2
}

# Mapeamento das disciplinas que cada professor pode dar
habilidades_professor = {
    "Prof_A": ["Matemática"],
    "Prof_B": ["Português", "História"],
    "Prof_C": ["Matemática", "Português"]
}

# Disponibilidade do professor (quais horários ele pode dar aula)
disponibilidade_professor = {
    "Prof_A": ["07:00", "08:00"],
    "Prof_B": ["08:00", "09:00"],
    "Prof_C": ["07:00", "09:00"]
}

# Criar cromossomo inicial aleatório respeitando carga horária
cromossomo = np.zeros((len(disciplinas), len(professores), len(horarios), len(turmas), len(dias)), dtype=int)

for d, disciplina in enumerate(disciplinas):
    total_aulas = carga_horaria_disciplina[disciplina]
    while total_aulas > 0:
        t = random.randint(0, len(turmas) - 1)  # Escolhe uma turma
        s = random.randint(0, len(dias) - 1)  # Escolhe um dia
        h = random.randint(0, len(horarios) - 1)  # Escolhe um horário
        p = random.choice(professores)  # Escolhe um professor aleatório

        if disciplina in habilidades_professor[p] and horarios[h] in disponibilidade_professor[p]:
            p_idx = professores.index(p)
            if cromossomo[d, p_idx, h, t, s] == 0:  # Evita dupla alocação
                cromossomo[d, p_idx, h, t, s] = 1
                total_aulas -= 1

# Print do Genótipo (Cromossomo como array)
print("Genótipo (Matriz de Alocações):")
print(cromossomo)

# Print do Fenótipo (Grade Horária Formatada)
print("\nFenótipo (Grade Horária Gerada):")
for s in range(len(dias)):
    print(f"\n--- {dias[s]} ---")
    for t in range(len(turmas)):
        print(f"\n{turmas[t]}")
        for h in range(len(horarios)):
            for p in range(len(professores)):
                for d in range(len(disciplinas)):
                    if cromossomo[d, p, h, t, s] == 1:
                        print(f"  {horarios[h]} - {disciplinas[d]} ({professores[p]})")
```
