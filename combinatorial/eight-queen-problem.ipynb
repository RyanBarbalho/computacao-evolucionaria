```python
import random
import numpy as np
import matplotlib.pyplot as plt
from random import choice
```

representação será uma lita de 8 elementos, onde cada elemento representa a coluna da rainha da linha i;
de inicio, não é possivel ter uma rainha na mesma linha, e por ter 8 elementos distintos, não é possivel ter uma rainha na mesma coluna;
ex: [1,3,5,7,2,4,8,6]

```python
TAMANHO_POPULACAO = 100
GERACOES = 1000
TAXA_MUTACAO = 0.8
TAXA_CRUZAMENTO = 1

def populacao_inicial():
	populacao = []
	for _ in range(TAMANHO_POPULACAO):
		individuo = list(range(0,8))
		random.shuffle(individuo)
		populacao.append(individuo)
	return populacao

#quanto menor o fitness, melhor o individuo
def funcao_fitness(individuo):
	fitness = 0
	for i in range(len(individuo)):
		for j in range(i+1, len(individuo)):
			if individuo[i]-individuo[j] == i - j or individuo[j]-individuo[i] == i-j:
				fitness +=1
	return -fitness + 8

def crossover(individuo1,individuo2):
	ponto_corte = random.randint(0,len(individuo1)-1)
	filho1 = individuo1[:ponto_corte]
	#previne numeros repetidos no crossover
	for i in range(0,len(individuo1)):
		if(individuo2[i] not in filho1):
			filho1.append(individuo2[i])
	filho2 = individuo2[:ponto_corte]
	for i in range(0, len(individuo2)):
		if(individuo1[i] not in filho2):
			filho2.append(individuo1[i])

	return filho1,filho2

def mutation(individuo, taxa_mutacao):
	if(random.randint(0,100) < taxa_mutacao*100):
		ponto_a = random.randint(0,7)
		ponto_b =  random.randint(0,7)
		while ponto_a == ponto_b:
			ponto_b = random.randint(0,7)
		aux = individuo[ponto_a]
		individuo[ponto_a] = individuo[ponto_b]
		individuo[ponto_b] = aux
	return individuo

#seleção será por torneio, k=5
#populacao and scores are lists
def tournament(populacao, scores,k=5):
	indices_selecionados = set()
	while len(indices_selecionados) < k:
		indices_selecionados.add(random.randint(0,len(populacao)-1))
	selecionados = [(populacao[i],scores[i]) for i in indices_selecionados]

	return max(selecionados, key=lambda x: x[1])[0]

def genetic_algorithm(taxa_mutacao, taxa_cruzamento, num_geracoes, tamanho_populacao):
	populacao = populacao_inicial()
	scores = []
	for _ in range(num_geracoes):
		nova_populacao = []
		scores = [funcao_fitness(individuo) for individuo in populacao]



		while len(nova_populacao) < tamanho_populacao:
			individuo1 = tournament(populacao, scores)
			individuo2 = tournament(populacao, scores)
			if(random.randint(0,1) < taxa_cruzamento):
				filho1,filho2 = crossover(individuo1,individuo2)
				filho1 = mutation(filho1, taxa_mutacao)
				filho2 = mutation(filho2, taxa_mutacao)
				nova_populacao.append(filho1)
				nova_populacao.append(filho2)
			else:
				nova_populacao.append(individuo1)
				nova_populacao.append(individuo2)

		populacao = nova_populacao

	populacao_final = sorted(populacao, key=lambda x: funcao_fitness(x), reverse=True)
	sorted_scores = sorted(scores, reverse=True)
	populacao_classificada = list(zip(populacao_final,sorted_scores))

	return populacao_classificada

populacao_encontrada = genetic_algorithm(TAXA_MUTACAO, TAXA_CRUZAMENTO, GERACOES, TAMANHO_POPULACAO)
print("Solucao encontrada:", populacao_encontrada[0][0])
print("score da solucao encontrada:", populacao_encontrada[0][1])







```
